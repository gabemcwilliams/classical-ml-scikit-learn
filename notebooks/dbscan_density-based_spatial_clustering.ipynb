```python
"""
DBSCAN vs KMeans Clustering Comparison

This notebook demonstrates basic clustering behavior on synthetic datasets
(blobs, moons, and circles) using both KMeans and DBSCAN algorithms.

It highlights differences in clustering performance based on dataset geometry,
showing where traditional clustering (KMeans) struggles compared to density-based clustering (DBSCAN).

Datasets:
- cluster_blobs.csv
- cluster_moons.csv
- cluster_circles.csv

Key Tools:
- Scikit-learn (KMeans, DBSCAN)
- Seaborn for visualization

"""

import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
blobs = pd.read_csv('/mnt/ml/mnt/mls/data/udemy/udemy/cluster_blobs.csv', dtype_backend='pyarrow')
```

```python
sns.scatterplot(x='X1', y='X2', data=blobs)
plt.show()
```

```python
moons = pd.read_csv('/mnt/ml/mnt/mls/data/udemy/udemy/cluster_moons.csv', dtype_backend='pyarrow')
```

```python
moons.head()
```

```python
sns.scatterplot(x='X1', y='X2', data=moons)
```

```python
circles = pd.read_csv('/mnt/ml/mnt/mls/data/udemy/udemy/cluster_circles.csv', dtype_backend='pyarrow')
```

```python
circles.head()
```

```python
sns.scatterplot(x='X1', y='X2', data=circles)
```

```python
def display_categories(model, data):
    labels = model.fit_predict(data)
    sns.scatterplot(x='X1', y='X2', hue=labels, data=data, palette='Set1')
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)

display_categories(model, blobs)
```

```python
display_categories(model, moons)
```

```python
display_categories(model, circles)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN()

display_categories(model, blobs)
```

```python
model = DBSCAN(
eps=0.15
)
```

```python
display_categories(model, moons)
```

## Summary

This simple comparison demonstrates that:

- **KMeans** performs well on spherical, evenly sized clusters but struggles with irregular shapes.
- **DBSCAN** excels at finding clusters of arbitrary shape and can effectively separate non-linear boundaries.

Choosing the right clustering algorithm depends heavily on the underlying geometry and noise level of the data.
